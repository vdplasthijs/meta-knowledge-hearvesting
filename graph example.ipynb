```python
import os, sys
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fp_entities = 'entity_type_map.txt'

with open(fp_entities, 'r') as f:
    lines = f.readlines()
    lines = [line.strip().split('\t') for line in lines]
```

```python
dict_ds = {}
list_ds = []
all_ent_types = []
'''

data structurte:

 ['=================================================='],
 ['Source URL: https://lifesciences.datastations.nl/dataset.xhtml?persistentId=doi:10.17026/dans-2bd-kskz'],
 [''],
 ['=================================================='],
 ['Metadata date:'],
 ['- (Dutch forest reserves database and network, 2019-04-11)'],
 etc


 '''

for line in lines:
    if line[0].startswith('Source URL:'):
        source_url = line[0].split(' ')[-1]
        dict_ds[source_url] = {}
        print(source_url)
    elif len(line[0]) == 0 or line[0].startswith('======================='):
        pass
    elif not line[0].startswith('- '):
        assert len(line) == 1
        ent_type = line[0].rstrip(':')
        dict_ds[source_url][ent_type] = {}
        all_ent_types.append(ent_type)
    else:
        line = line[0]
        line = line.lstrip('- (').rstrip(')')
        line = line.split(',')
        ent_name = line[0]
        ent_descr = line[1:]
        dict_ds[source_url][ent_type][ent_name] = ent_descr
        # assert len(dict_ds[source_url][ent_type]) == 1, f"multiple entities with the different names in {source_url}: {dict_ds[source_url][ent_type]}"

all_ent_types = list(set(all_ent_types))
```

```python
all_ent_types
```

```python
ent_types_keep = ['Title', 'Keywords', 'Data creator', 'Spatial coverage', 'Temporal coverage', 'Description']
ent_types_keep = [et for et in all_ent_types if et in ent_types_keep]

df = pd.DataFrame(columns=['source_url', 'entity_type', 'entity_name', 'entity_description'])
for source_url, ent_types in dict_ds.items():
    for ent_type, ent_names in ent_types.items():
        for ent_name, ent_descr in ent_names.items():
            if ent_type in ent_types_keep:
                df = pd.concat([df, pd.DataFrame([{'source_url': source_url,
                                                   'entity_type': ent_type,
                                                   'entity_name': ent_name,
                                                   'entity_description': ', '.join(ent_descr)}])], ignore_index=True)
df[:30]
```
